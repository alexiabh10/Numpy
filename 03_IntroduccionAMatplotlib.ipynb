uso de Python.

## Import

## Representación gráfica de datos

Si ala función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical y utilizará el índice de cada punto de datos en el Array como la coordenada horizontal.


```python
plt.plot([1,2,5,7,8,1])
plt.show()
```

```python
# Tambien se puede proporcionar 2 matrices; una para el eje horozontal y otra para el eje vertical
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.show
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan saturada.

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.axis([-4,8,0,10]) # [xmin,xmax,ymin,ymax]
plt.show
```

Se sigue el mismo procedimiento para pintar una función matemática.

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
plt.plot(x,y)
plt.show
```

Tambien puede modificars ek estilo de la gráfica para que contenga más información

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1

plt.plot(x,y, 'b--',x,y2, 'g')
plt.show
```

```python
# Separando en diferentes lineas las funciones
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1

plt.plot(x,y, 'b--')
plt.plot(x,y2,'g')
plt.show
```

Para poder diferenciar entre ambas funciones siempres es recomendable añadir una leyenda

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1
plt.plot(x,y,'b--',label = "x**2")
plt.plot(x,y2,'g',label = "x + 1")
plt.legend(loc = "best") # La situa en la mejor localización
plt.show
```

Tambien pueden crearse dos gráficas que no se superpongan. Estas gráficas se organizan en un **grid** y se denominan **subplots**.

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1

plt.subplot(1,2,1) # 1 es una fila(rows), 1Columnas(columns), 1 primer subplot
plt.plot(x,y,'b--')


plt.subplot(1,2,2) # 1 es una fila(rows), 2 Columnas(columns), 2 segundo subplot
plt.plot(x,y2,'g')

plt.show
```

Para que las gráfcas no queden tan ajustadas, podemos haccer la figura más grande.

```python
plt.figure(figsize = (14,6))

plt.subplot(1,2,1) # 1 es una fila(rows), 1Columnas(columns), 1 primer subplot
plt.plot(x,y,'b--')


plt.subplot(1,2,2) # 1 es una fila(rows), 2 Columnas(columns), 2 segundo subplot
plt.plot(x,y2,'g')

plt.show
```

## Satter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x,y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.show
```

## Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211)# El primer numero indica 2 filas, 1 columna, 1 subplot
plt.hist(data,bins = 10,rwidth = 0.8) #bins=alto rwidth=ancho

plt.subplot(212) # 2 Rows, 1 Columns, 2nd Subplot
plt.hist(data,bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

## Guardar Figuras

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.plot(x,y, 'b--', label = "x**2")
plt.plot(x,y2, 'g',label = "x+1")
plt.legend(loc = "best")
plt.savefig("Mi_Grafiquita.png", transparent = True)
```
